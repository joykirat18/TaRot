```python
import torch
import pickle

def load_checkpoint(filepath):
    # Load the checkpoint dictionary
    with open(filepath, 'rb') as f:
        checkpoint = pickle.load(f)
    
    train_X = checkpoint['train_X']
    train_Y = checkpoint['train_Y']
    gp_state_dict = checkpoint['gp_state_dict']
    iteration = checkpoint['iteration']
    
    return train_X, train_Y, gp_state_dict, iteration
```

```python
filePath = "BaysianOptimization/checkpoint/phi-3/SNLI/NN_kernel_layer_0_6_angle_-0.5235987755982988_0.5235987755982988_reasoning_prob_v2.pkl"
```

```python
train_X, train_Y, gp_state_dict, iteration = load_checkpoint(filePath)
```

```python
y = []
for i in range(len(train_X)):
    y.append(train_Y[i].item())
```

```python
import matplotlib.pyplot as plt
import numpy as np
# train_Y is y axis
# iterations is x axis
plt.plot(np.arange(len(y)), y)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss vs Iterations')
```
